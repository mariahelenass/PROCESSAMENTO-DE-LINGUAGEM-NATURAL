```python
# Importando uma pagina da WEB
import nltk
```

```python
from urllib.request import urlopen
```

```python
url = "https://pt.wikipedia.org/wiki/Travis_Scott"
```

```python
html = urlopen(url).read()
```

```python
html
```

```python
from bs4 import BeautifulSoup
```

```python
web_str = BeautifulSoup(html).get_text()
```

```python
web_str
```

```python
start = web_str.find("Aos 17 anos, Travis iniciou sua carreira como produtor musical, com foco na produção de hip-hop.")
```

```python
start
```

```python
end = web_str.find("Em 2008, a dupla lançou o seu primeiro EP sem título na rede social Myspace.")
```

```python
last_sent = len("Em 2008, a dupla lançou o seu primeiro EP sem título na rede social Myspace.")
```

```python
intro = web_str[start:end+last_sent]
```

```python
intro
```

```python
intro_tokens = nltk.word_tokenize(intro)
```

```python
intro_tokens
```
